# Menentukan K (n_cluster) tanpa label

### Menggunakan inertia / Elbow Analysis

```python
X = X1
score = []

for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X, y=None) 

    score.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, 'r-')
plt.ylabel('Inertia')
plt.xlabel('n_cluster')
```

<b> n_cluster = 4 </b> <br>
cari yang melandai atau sikunya (yang paling bengkok)

clustering yang baik: <br>
jarak antara member dengan centroid itu dekat <br>
jarak antara centroid itu jauh

# Menentukan K (n_cluster) dan ada label (target)

### Menggunakan Silhoutte Score

```python
from sklearn.metrics import silhouette_score
```

```python
X = X1
score = []

for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(X, y=None) 

    silhoutte = silhouette_score(X, member, metric='euclidean')
    score.append(silhoutte)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, 'r-')
plt.ylabel('silhoutte')
plt.xlabel('n_cluster')
```

<b> n_cluster = 4 </b> <br>
cari yang memiliki nilai score tertinggi
